# Importing Modules and the Dataset

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('output.csv')
print(df)
```

# Data Analysis

```python
n_registers = df.shape[0]
print('Number of registers:', n_registers)
n_variables = df.shape[1]
print('Number of Variables', n_variables)
```

```python
#The occurencies of empty values in the given dataset is marked with '?'

empty_value = '?'
res = 0
for column in df.columns:
    occurrences = df[column].value_counts().get(empty_value, 0)
    res += occurrences

print('Number of empty values:', res )
```

```python
for column in df.columns[:-1]:
    print('============',column, '============')
    print('Min value: ', df[column].min())
    print('Max value: ', df[column].max())
    print('Mean value: ', df[column].mean())
    print('Std deviation value: ', df[column].std())
    print('Variance value: ', df[column].var())
    print('============', column, '============\n')
```

```python
for column in df.columns[:-1]:
    sns.histplot(data=df, x=column)
    plt.show()
```

```python
sns.displot(data=df, x="mpg", kde=True)
```

```python
#Principal Components Analysis

X = df.drop('car name', axis=1)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA()
pca.fit(X_scaled)

explained_variance_ratio = pca.explained_variance_ratio_
principal_components = pca.components_

print("Explained Variance Ratio:", explained_variance_ratio)
print("Principal Components:", principal_components)
```
